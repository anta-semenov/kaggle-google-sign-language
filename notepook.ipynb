```python
def get_unique_frames(path):
    df = pd.read_parquet('asl-signs/' + path)
    return df['frame'].nunique(), df[(df['type'] == 'left_hand') & (df['x'].isna() == False)].sum(), df[(df['type'] == 'right_hand') & (df['x'].isna() == False)].sum()

df = pd.read_csv('asl-signs/train.csv')

# take subset of df
df = df.iloc[0:10]
df['amount_of_frames'], df['left_hand_usage'], df['right_hand_usage'] = df['path'].apply(get_unique_frames)

df.head(10)
```
